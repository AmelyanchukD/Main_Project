# RFM анализ

<span style="font-weight: bold;">Для RFM анализа удаляем строки, по которым осуществлялся возврат</span>

```python
filtered_df = df[df['returnflag'] == False].drop('productname', axis=1)
filtered_df.head()
```

```python
# Преобразуем столбец 'date' в формат datetime
filtered_df['date'] = pd.to_datetime(filtered_df['date'])

# Рассчитываем Recency
latest_date = filtered_df['date'].max()
recency_data = filtered_df.groupby('customerid')['date'].max().reset_index()
recency_data['recency'] = (latest_date - recency_data['date']).dt.days

# Рассчитываем Frequency
frequency_data = filtered_df.groupby('customerid').size().reset_index(name='frequency')

# Рассчитываем Monetary
monetary_data = filtered_df.groupby('customerid')['totalprice'].sum().reset_index()

# Объединяем данные
rfm_data = recency_data.merge(frequency_data, on='customerid').merge(monetary_data, on='customerid')

recency_quintiles = pd.qcut(rfm_data['recency'], 4, labels=False)
rfm_data['R'] = recency_quintiles + 1  # +1 для того, чтобы рейтинги начинались с 1

# Добавляем квинтили для Frequency
frequency_quintiles = pd.qcut(rfm_data['frequency'], 4, labels=False)
rfm_data['F'] = frequency_quintiles + 1  # +1 для того, чтобы рейтинги начинались с 1

# Добавляем квинтили для Monetary
monetary_quintiles = pd.qcut(rfm_data['totalprice'], 4, labels=False)
rfm_data['M'] = monetary_quintiles + 1  # +1 для того, чтобы рейтинги начинались с 1

# Создаем столбец 'RFMClass', объединяя значения рейтингов в одну строку
rfm_data['RFMClass'] = rfm_data['R'].astype(str) + rfm_data['F'].astype(str) + rfm_data['M'].astype(str)

# Выводим результат
print(rfm_data)
```

```python
import matplotlib.pyplot as plt

# Построение гистограммы распределения RFM-классов
plt.figure(figsize=(20, 6))
rfm_data['RFMClass'].value_counts().sort_index().plot(kind='bar', color='darkred')
plt.title('Distribution of RFM Classes')
plt.xlabel('RFM Class')
plt.ylabel('Number of Customers')
plt.xticks(rotation=0)
plt.show()
```

# Retention Rate

```python
df_retention = df[['date', 'customerid']]
```

```python
df_retention['date'] = pd.to_datetime(df_retention['date'])
```

```python
#Возьмём дату, когда пользователь впервые проявил активность
first_session_date_by_customers = df_retention.groupby('customerid')['date'].min()

first_session_date_by_customers.name = 'first_session_date'
first_session_date_by_customers
```

```python
visits = df_retention.join(first_session_date_by_customers,on='customerid')
visits['cohort_lifetime'] = visits['date'] - visits['first_session_date']
visits
```

```python
#Сгруппируем данные по когорте и lifetime
cohorts = visits.groupby(['first_session_date', 'cohort_lifetime']).agg({'customerid': 'nunique'}).reset_index()
cohorts
```

```python
# Получаем количество пользователей в когорте на день первой сессии
initial_users_count = cohorts[cohorts['cohort_lifetime'] == pd.Timedelta(0)][['first_session_date', 'customerid']]
initial_users_count = initial_users_count.rename(columns={'customerid': 'cohort_users'})
```

```python
# Объединяем с когортами, чтобы добавить количество пользователей в когорте
cohorts = cohorts.merge(initial_users_count, on='first_session_date', how='left')
```

```python
# Рассчитываем коэффициент удержания
cohorts['retention'] = cohorts['customerid'] / cohorts['cohort_users']
```

```python
# Фильтруем данные для последних двух лет и cohort_lifetime равным 14 дням
start_date = pd.Timestamp('2018-11-01')
end_date = pd.Timestamp('2018-11-14')

filtered_cohorts = cohorts[(cohorts['first_session_date'] >= start_date) & 
                            (cohorts['first_session_date'] <= end_date) & 
                            (cohorts['cohort_lifetime'].dt.days >= 0) & 
                            (cohorts['cohort_lifetime'].dt.days <= 14)]
```

```python
# Создаем сводную таблицу
retention_pivot = filtered_cohorts.pivot_table(index='first_session_date', 
                                                columns='cohort_lifetime',
                                                values='retention', 
                                                aggfunc='sum')

# Заменяем пустые значения на NaN
retention_pivot.replace('', float('nan'), inplace=True)

# Проверяем результат
retention_pivot

```

```python
# Построение тепловой карты
sns.set(style='white')
plt.figure(figsize=(13, 9))
plt.title('Cohorts: User Retention')
sns.heatmap(retention_pivot, annot=True, fmt='.1%', linewidths=1, linecolor='gray', cmap='Greens');
plt.show()
```
